# Analyse de l’attrition des employés  
Ce notebook analyse les facteurs qui influencent le départ des employés dans une entreprise technologique.  
Les données utilisées proviennent de plusieurs fichiers CSV fournis.  


```python
# analysis.ipynb

# 1. Import des bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Afficher les graphiques dans le notebook
%matplotlib inline

# 2. Chargement des fichiers CSV
employee = pd.read_csv("employee_survey_data.csv")
manager = pd.read_csv("manager_survey_data.csv")
general = pd.read_csv("general_data.csv")
in_time = pd.read_csv("in_time.csv")
out_time = pd.read_csv("out_time.csv")

# 3. Aperçu des données
print("employee_survey_data:")
display(employee.head())

print("manager_survey_data:")
display(manager.head())

print("general_data:")
display(general.head())

# 4. Fusion des données sur EmployeeID
df = general.merge(employee, on='EmployeeID') \
            .merge(manager, on='EmployeeID')

print(f"Dimensions après fusion : {df.shape}")

# 5. Nettoyage des données : vérifier valeurs manquantes
print("Valeurs manquantes par colonne :")
print(df.isnull().sum())

# Supprimer les lignes avec valeurs manquantes
df = df.dropna()
print(f"Dimensions après suppression des valeurs manquantes : {df.shape}")

# 6. Statistiques descriptives des variables numériques
print("Statistiques descriptives :")
display(df.describe())

# 7. Visualisation : Distribution d'une variable numérique
plt.figure(figsize=(8,4))
sns.histplot(df['EnvironmentSatisfaction'], bins=5, kde=False)
plt.title("Distribution de EnvironmentSatisfaction")
plt.show()

# 8. Visualisation : Boxplot pour attrition vs salaire
plt.figure(figsize=(8,4))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Salaire mensuel selon l'attrition")
plt.show()

# 9. Matrice de corrélation
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Matrice de corrélation")
plt.show()

# 10. Préparation des données pour modélisation

# Encodage de la variable cible (Attrition)
df['Attrition'] = df['Attrition'].map({'Yes':1, 'No':0})

# Encodage des variables catégorielles (excluant EmployeeID, attrition déjà encodée)
df_encoded = pd.get_dummies(df.drop(['EmployeeID'], axis=1), drop_first=True)

# Séparation des variables explicatives et cible
X = df_encoded.drop('Attrition', axis=1)
y = df_encoded['Attrition']

# Normalisation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Séparation train/test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 11. Modélisation avec régression logistique
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# 12. Évaluation du modèle
print("Matrice de confusion :")
print(confusion_matrix(y_test, y_pred))

print("\nRapport de classification :")
print(classification_report(y_test, y_pred))

# 13. Interprétation des coefficients
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
}).sort_values(by='Coefficient', ascending=False)

print("Coefficients du modèle (importance des variables) :")
display(coefficients)

```

## Résumé et recommandations

- Les variables ayant les coefficients les plus élevés impactent le plus l’attrition.
- Par exemple, `OverTime_Yes` a un coefficient positif élevé, ce qui signifie que les employés en heures supplémentaires ont plus de chances de quitter.
- La satisfaction au travail, l'ancienneté et le nombre de projets sont aussi des facteurs clés.
- Recommandations : cibler les employés travaillant beaucoup d'heures supplémentaires pour améliorer la rétention, renforcer la satisfaction au travail.
- Le modèle montre une bonne précision pour prédire l’attrition et peut aider la RH à prendre des décisions.
